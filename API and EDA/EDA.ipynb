# EDA

```python
import numpy as np
import pandas as pd
import geopandas as gpd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
import libpysal as ps
from shapely.geometry import Point
```

```python
sh = gpd.read_file("shp/---polygon.shp") 
shp=gpd.read_file("shp/深圳各街道乡镇.shp") 
m = pd.read_csv("output_data/metro.csv")
h =pd.read_csv("output_data/hospital.csv")
i =pd.read_csv("output_data/industry.csv")
b =pd.read_csv("output_data/building.csv")
f =pd.read_csv("output_data/farm.csv")
s =pd.read_csv("output_data/shop.csv")
p =pd.read_csv("output_data/park.csv")
pop=pd.read_csv("population.csv")
```

```python
crs = {'init': 'epsg:4326'}
shp1 = shp.to_crs("EPSG:4326")
```

## Metro

```python
#Transfrom lng and lat to geometry with epsg4326
geometry = [Point(xy) for xy in zip(m['lng'], m['lat'])]
mgdf = gpd.GeoDataFrame(m, crs=crs, geometry=geometry)
```

```python
#plot
ax = shp1.plot(cmap = 'jet', column = 'Layer', figsize=(10,10)) 
mgdf.plot(ax=ax,color='yellow') 
```

## Hospital

```python
#Transfrom lng and lat to geometry with epsg4326
geometry = [Point(xy) for xy in zip(h['lng'], h['lat'])]
hgdf = gpd.GeoDataFrame(h, crs=crs, geometry=geometry)
```

## Industry
