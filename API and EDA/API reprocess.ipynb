# API Obtain via Baidu map

```python
from urllib import request
from urllib.parse import quote
import json
import string
from pandas.core.frame import DataFrame
```

## API of metro station in Shenzhen

```python
subwayList = []
resultss_name=[]
resultss_lat=[]
resultss_lng=[]
resultss_address=[]
with open('metro_name', 'r') as f:
    subwayList=f.read().strip(',').split('\n')

def startGet():
    for emp in subwayList:
        print(emp)
        getLocation(address=emp)

def getLocation(address='address'):
    url = "http://api.map.baidu.com/place/v2/search?query="+address+"&tag=华强本站&region=深圳市&output=json&ak=xnrnvAZtT68KPGaw5MRcHkhMPlf8hAFi"
    url = quote(url, safe=string.printable)
    req = request.urlopen(url)
    res = req.read().decode('utf-8','ignore').replace(u'\xa9', u'')
    result = json.loads(res)
    print(result)
    
    result233=[]
    result233.append(result)
    results=[]
    results.append(result233[0])
    resultss_name.append(results[0]['results'][0]['name'])
    resultss_lat.append(results[0]['results'][0]['location']['lat'])
    resultss_lng.append(results[0]['results'][0]['location']['lng'])
    resultss_address.append(results[0]['results'][0]['address'])

if __name__ == '__main__':
    startGet()
```

```python
c={'name':resultss_name,
  "lat":resultss_lat,
  "lng":resultss_lng,
  "district":resultss_address}
metro_data=DataFrame(c)
print(metro_data)
```

```python
metro_data=metro_data.set_index('name')
```

```python
metro_data.to_csv("output_data/metro.csv")
```
